```python
# Ø­Ø°Ù Ø£ÙŠ ØµÙ ÙÙŠÙ‡ Ù‚ÙŠÙ… Ù†Ø§Ù‚ØµØ©
df = df.dropna(subset=["comment", "label"])
df = df[df["label"].str.strip() != ""]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

# Ù‚Ø±Ø§Ø¡Ø© Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
df = pd.read_csv(r"C:\Users\infor\OneDrive\Ø³Ø·Ø­ Ø§Ù„Ù…ÙƒØªØ¨\DB\Tawacol_youtube_labeled_with_labels.csv")

# Ù†Ø³ØªØ®Ø¯Ù… Ø§Ù„Ù†ØµÙˆØµ Ø§Ù„Ù†Ø¸ÙŠÙØ©
X = df["comment"]
y = df["label"]
```

```python

#  ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ù†ØµÙˆØµ Ø§Ù„Ù‰ Ø§Ø±Ù‚Ø§Ù… 
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_vectorized = vectorizer.fit_transform(X)

# ØªÙ‚Ø³ÙŠÙ… Ø§Ù„Ø¨ÙŠØ§Ø§Ù†Ø§Øª
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42
)

# ØªØ¯Ø±ÙŠØ¨
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±
y_pred = model.predict(X_test)

print("ğŸ“Š ØªÙ‚Ø±ÙŠØ± Ø§Ù„ØªØµÙ†ÙŠÙ:")
print(classification_report(y_test, y_pred, digits=3))

print("\nğŸŒ€ Ù…ØµÙÙˆÙØ© Ø§Ù„Ø§Ù„ØªØ¨Ø§Ø³:")
print(confusion_matrix(y_test, y_pred))
```
