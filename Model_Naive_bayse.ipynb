```python
#أكرم محمد محمد علي عقلان
#من الافضل رونالدو ام ميسي
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# رابط الفيديو
video_url = "https://youtu.be/q-3S3rlDvN4?si=IIiJZ2GXyWWO23Yu"

driver = webdriver.Chrome()
driver.get(video_url)

time.sleep(5)

# تمرير لأسفل لجلب المزيد من التعليقات
for _ in range(5):
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    time.sleep(3)

# جلب التعليقات عبر span
comments = driver.find_elements(
    By.XPATH, '//span[contains(@class, "yt-core-attributed-string")]'
)

data = []
for c in comments:
    text = c.text.strip()
    if text:  # تجاهل التعليقات الفارغة
        data.append(text)

df = pd.DataFrame(data, columns=["comment"])
df.to_csv("youtube_comments.csv", index=False, encoding="utf-8-sig")

driver.quit()
print("✅ تم حفظ التعليقات باستخدام span class في youtube_comments.csv")
```

```python

# تحميل الملف
df = pd.read_csv(r"C:\Users\infor\OneDrive\سطح المكتب\DB\Tawacol.csv")

# دالة لتصنيف التعليق مبدئيا بالكلمات المفتاحية
def simple_label(text):
    text = str(text)
    if any(word in text for word in ["رونالدو","رونالدوا افضل لاعب ","رونالدو الافضل","رونالدو افضل من ميسي","رونالدو عم ميسي","كريستيانو الاسطورة","كريستيانو رونالدو"]):
        return "ايجابي"
    elif any(word in text for word in [ "ميسي افضل","ميسي عم كريستيانو","ميسي افضل لاعب في العالم","ميسي","ميسي الاسطورة","ميسي الافضل","ميسي رائع","لونيل ميسي","ميسي افضل من لمس الكرة"]):
        return "سلبي"
    else:
        return "محايد"

# إضافة العمود
df["label"] = df["comment"].apply(simple_label)

# حفظ الملف الجديد
df.to_csv(r"C:\Users\infor\OneDrive\سطح المكتب\DB\Tawacol_youtube_labeled_with_labels.csv", index=False, encoding="utf-8-sig")
print("✅ تم إنشاء عمود label وحفظه في youtube_labeled_with_labels.csv")
```

```python
# حذف أي صف فيه قيم ناقصة
df = df.dropna(subset=["comment", "label"])
df = df[df["label"].str.strip() != ""]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

# قراءة البيانات
df = pd.read_csv(r"C:\Users\infor\OneDrive\سطح المكتب\DB\Tawacol_youtube_labeled_with_labels.csv")

# نستخدم النصوص النظيفة
X = df["comment"]
y = df["label"]
```

```python

#  تحويل النصوص الى ارقام 
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_vectorized = vectorizer.fit_transform(X)

# تقسيم البياانات
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42
)

# تدريب
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# الاختبار
y_pred = model.predict(X_test)

print("📊 تقرير التصنيف:")
print(classification_report(y_test, y_pred, digits=3))

print("\n🌀 مصفوفة الالتباس:")
print(confusion_matrix(y_test, y_pred))
```

```python
test_comment = ["كرة القدم"]
test_comment_clean = [test_comment[0]]
test_vectorized = vectorizer.transform(test_comment_clean)

prediction = model.predict(test_vectorized)
print("🔮 التصنيف:", prediction[0])
```
